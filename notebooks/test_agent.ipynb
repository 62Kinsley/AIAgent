import sys
sys.path.append("..")
from app.agents.personal_chief import agent

from langchain_core.messages import HumanMessage

multimodal_message = HumanMessage(
    content=[
        {
            "type": "image_url",
            "image_url": {"url": "https://img.freepik.com/free-photo/arrangement-different-foods-organized-fridge_23-2149099882.jpg"}
        },
        {
            "type": "text",
            "text": "帮我看看这些食材能做些什么？"
        }
    ]
)

config = {"configurable": {"thread_id": "6"}}

response = agent.invoke({"messages": [multimodal_message]}, config)

for message in response["messages"]:
    message.pretty_print()
